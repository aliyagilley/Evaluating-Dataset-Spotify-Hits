```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plot
from matplotlib import pyplot as plot
df = pd.read_csv("top10s.csv", encoding = "ISO-8859-1", engine = "python")
df
```

```python
#20 artists with the most hits
df["artist"].value_counts().head(20).plot.bar()
```

```python
#want to know the top 10 popular artists from 2010-2019 from best to least best. Let's see if I can figure out how...
#this will be  different than seeing what artists have the "most" hits; this will be looking at the most popular hits.
#finding the average popularity of each artist and putting in ascending order
df_avg_pop_artist = df[['artist','pop']].groupby('artist').mean().astype(int).reset_index().sort_values(by='pop', ascending = False, inplace = False).head(10)

sns.catplot(x = 'artist', y = 'pop', data = df_avg_pop_artist, kind ='bar', aspect = 3)
plot.xlabel('Artist')
plot.ylabel('Popularity')
plot.title('Top 10 Popular Artists')
```

```python
#thought it would be interesting to see the least popular artists from this decade!  Let's see if any of your favorites (or least favorites) are in here.
df_avg_least_pop_artist = df[['artist','pop']].groupby('artist').mean().astype(int).reset_index().sort_values(by='pop', ascending = False, inplace = False).tail(10)

sns.catplot(x = 'artist',y = 'pop', data = df_avg_least_pop_artist, kind ='bar', aspect = 3)
plot.xlabel('Artist')
plot.ylabel('Popularity')
plot.title('Top 10 Least Popular Artists')
```

```python
#want to know the top 10 popular songs from 2010-2019. Let's see if I can figure out how...

df_avg_pop_song = df[['title','pop']].groupby('title').mean().astype(int).reset_index().sort_values(by='pop', ascending = False, inplace = False).head(10)
#a friend that is much better at python told me about catplot, and it's super useful! Helps me to visualize the data more.
sns.catplot(x = 'title',y = 'pop', data = df_avg_pop_song, kind ='bar', aspect = 8.5)
plot.xticks(fontsize = 15)
plot.xlabel('Song Title', fontsize = 15)
plot.ylabel('Popularity', fontsize = 15)
plot.title('Top 10 Popular Songs')
```

```python
# Run the simulation for liveness 10000 times
#need to create a function to simulate average first
def simulate(column, n):
    return df[column].sample(n).mean()
#now to simulate
data = []
for i in range(1000):
    live100 = simulate("live", 100)
    d = {"live100" : live100}
    data.append(d)

df_live_100 = pd.DataFrame(data)

#create hist
df_live_100.hist()
```
